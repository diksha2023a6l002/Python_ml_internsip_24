```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
auto=pd.read_csv("Automobile_data.csv")
auto.head()
```

```python
plt.plot(auto["make"],auto["fuel"])
plt.show()
```

```python
plt.bar(auto["make"],auto["fuel-type"])
plt.show()
```

```python
brandVSfuel=auto.groupby(["make", "fuel-type"]).size().unstack(fill_value=0)
brandVSfuel
```

```python
brandVSfuel.plot(kind="bar", stacked=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))

sns.lineplot(data=auto["wheel-base"], linewidth=1, label="Wheel-Base")
sns.lineplot(data=auto["engine-size"], linewidth=1, label="Engine-Size")

plt.title("Wheel Base VS Engine Size", color='y')
plt.xlabel("Engine Size", color='g', fontsize=15.5)
plt.ylabel("Wheel Base", color='r', fontsize=15.5)
plt.show
```

```python
plt.figure(figsize=(20,8))
sns.set(rc={"axes.facecolor":"#FFFF0F", "axes.grid":False, "xtick.labelsize":10, "ytick.labelsize":10})

plt.title("Wheel Base VS Engine Size")
plt.ylabel("Wheel Base")
plt.xlabel("Engine Size")
plt.xticks(rotation=45)
sns.lineplot(x=auto["engine-size"],y=auto["wheel-base"].index.values,color='r')
plt.show()
```
